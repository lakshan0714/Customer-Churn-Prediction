```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras import Sequential,layers
from tensorflow.keras.callbacks import TensorBoard
import datetime
import numpy as np
```

```python
df=pd.read_csv('Churn.csv')
df.head()
df.info()
```

```python
df.drop("customerID",inplace=True,axis='columns')
```

```python
df.info()
```

```python
df['TotalCharges'].dropna()
```

```python

df1=df[df.TotalCharges!=' ']
df1.TotalCharges=pd.to_numeric(df1['TotalCharges'])
```

```python
df1.info()
```

```python
df1.Churn
```

```python
def print_unique(df):
    for column in df:
       if df[column].dtype=='object':
           print(f'{column}:{df1[column].unique()}')
```

```python
print_unique(df1)
```

```python
df1.replace('No phone service','No',inplace=True)
df1.replace('No internet service','No',inplace=True)
```

```python
df1.info()
```

```python
df1.replace({'Yes':1,'No':0},inplace=True)
df1.replace({True:1,False:0},inplace=True)

df1.replace({'Male':1,'Female':0},inplace=True)
```

```python
df1.info()
```

```python
df1.head()
```

```python
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
df2.replace({True:1,False:0},inplace=True)
df2.info()
```

Creating Neural NEtwork(ANN)

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
X=df2.drop('Churn',axis='columns')
Y=df2['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_Train,X_test,Y_Train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=True,shuffle=True)
X_Train.shape

X_Train.head()
```

```python
model=Sequential([
    layers.Dense(256,input_shape=(26,),activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(128,input_shape=(26,),activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(64,activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(32,activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(16,activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
#log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
#tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

model.fit(X_Train,Y_Train,validation_split=0.2,epochs=200 )
```

```python
model.evaluate(X_test,Y_test)











from sklearn.metrics import confusion_matrix

```
